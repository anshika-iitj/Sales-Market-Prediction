**data pre processing**

```
# checking unique values in the columns of train dataset

data.apply(lambda x: len(x.unique()))
```

```
data.isnull().sum()
```

```
# imputing missing values

data['Item_Weight'] = data['Item_Weight'].replace(0, np.NaN)
data['Item_Weight'].fillna(data['Item_Weight'].mean(), inplace = True)

data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0], inplace = True)

data['Item_Outlet_Sales'] = data['Item_Outlet_Sales'].replace(0, np.NaN)
data['Item_Outlet_Sales'].fillna(data['Item_Outlet_Sales'].mode()[0], inplace = True)

data.isnull().sum()
```

```
# combining reg, Regular and Low Fat, low fat and, LF


data['Item_Fat_Content'] = data['Item_Fat_Content'].replace({'LF': 'Low Fat', 'reg': 'Regular', 'low fat': 'Low Fat'})
data['Item_Fat_Content'].value_counts()
```

```
# Getting the first two characters of ID to separate them into different categories

data['Item_Identifier'] = data['Item_Identifier'].apply(lambda x: x[0:2])

data['Item_Identifier'] = data['Item_Identifier'].map({'FD':'Food', 'NC':'Non_Consumable', 'DR':'Drinks'})

data['Item_Identifier'].value_counts()
```

```
# determining the operation peroid of a time

data['Outlet_Years'] = 2013 - data['Outlet_Establishment_Year']
data['Outlet_Years'].value_counts()
```

```
# removing unnecassary columns from the dataset

#data = data.drop('Item_Identifier', axis = 1)

#print(data.shape)
```

```
data['Outlet_Type'].value_counts()
```

```
# label encoding

from sklearn.preprocessing import LabelEncoder

data.apply(LabelEncoder().fit_transform)
```

```
# one hot encoding

data = pd.get_dummies(data)

print(data.shape)
```

```
# splitting the data into dependent and independent variables

x = data.drop('Item_Outlet_Sales', axis = 1)
y = data.Item_Outlet_Sales

print(x.shape)
print(y.shape)
```

```
# splitting the dataset into train and test

train = data.iloc[:8523,:]
test = data.iloc[8523:,:]

print(train.shape)
print(test.shape)
```

```
# making x_train, x_test, y_train, y_test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

**Modelling**

**Linear Regression**

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(x_train, y_train)

# predicting the  test set results
y_pred = model.predict(x_test)
print(y_pred)

# finding the mean squared error and variance
mse = mean_squared_error(y_test, y_pred)
print('RMSE :', np.sqrt(mse))
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```

**AdaBoost Regressor**

```
from sklearn.ensemble import AdaBoostRegressor

model= AdaBoostRegressor(n_estimators = 100)
model.fit(x_train, y_train)

# predicting the test set results
y_pred = model.predict(x_test)

# RMSE
mse = mean_squared_error(y_test, y_pred)
print("RMSE :", np.sqrt(mse))
```

**XgBoost Regressor**

```
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor()
model.fit(x_train, y_train)

# predicting the test set results
y_pred = model.predict(x_test)
print(y_pred)

# Calculating the root mean squared error
print("RMSE :", np.sqrt(((y_test - y_pred)**2).sum()/len(y_test)))
```

**Random Forest Regression**

```
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators = 100 , n_jobs = -1)
model.fit(x_train, y_train)

# predicting the  test set results
y_pred = model.predict(x_test)
print(y_pred)

# finding the mean squared error and variance
mse = mean_squared_error(y_test, y_pred)
print("RMSE :",np.sqrt(mse))
print('Variance score: %.2f' % r2_score(y_test, y_pred))

print("Result :",model.score(x_train, y_train))
```

**Decision Tree Regressor**

```
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train, y_train)

# predicting the test set results
y_pred = model.predict(x_test)
print(y_pred)

print(" RMSE : " , np.sqrt(((y_test - y_pred)**2).sum()/len(y_test)))
```

**Support vector machine**

```
from sklearn.svm import SVR

model = SVR()
model.fit(x_train, y_train)

# predicting the x test results
y_pred = model.predict(x_test)

# Calculating the RMSE Score
mse = mean_squared_error(y_test, y_pred)
print("RMSE :", np.sqrt(mse))
```

```
import tensorflow as tf
import numpy as np

# Convert x_train and y_train to NumPy arrays with dtype float32
x_train = np.asarray(x_train).astype(np.float32)  # Convert to array and set dtype
y_train = np.asarray(y_train).astype(np.float32)  # Convert to array and set dtype

# Reshape y_train to have shape (number_of_samples, 1)
y_train = y_train.reshape(-1, 1)  # Reshape y_train before fitting the model


# Define the neural network model using Keras Functional API
D = x_train.shape[1]
tf_X = tf.keras.Input(shape=(D,)) # Use tf.keras.Input instead of placeholder

# Layer 1
Layer_1 = tf.keras.layers.Dense(20, activation='relu')(tf_X)

# Layer 2
Layer_2 = tf.keras.layers.Dense(15, activation='relu')(Layer_1)

# Layer 3
Layer_3 = tf.keras.layers.Dense(10, activation='relu')(Layer_2)

# Output layer
output = tf.keras.layers.Dense(1)(Layer_3)

# Create the model
model = tf.keras.Model(inputs=tf_X, outputs=output)

# Define cost function and optimizer
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss='mean_squared_error') # Use Adam optimizer

# Train the model
history = model.fit(x_train, y_train, epochs=100, verbose=1) # Store training history

# Visualization of training data
plt.plot(history.history['loss']) # Access loss from history
plt.title(" Training Costs vs Epochs ")
plt.xlabel(" Epochs ")
plt.ylabel(" Costs ")
plt.show()

# Evaluate on test data (convert x_test, y_test if needed)
x_test = np.asarray(x_test).astype(np.float32)
y_test = np.asarray(y_test).astype(np.float32)
y_test = y_test.reshape(-1, 1)

loss = model.evaluate(x_test, y_test, verbose=0)
print("Test Loss:", loss)
```
